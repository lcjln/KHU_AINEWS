## Perpelxity 이슈 발굴

```python
import requests
import datetime

# API 정보
API_KEY = "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"  # 본인의 API 키 입력
url = "https://api.perplexity.ai/chat/completions"

# 오늘 날짜 가져오기 (YYYY년 M월 D일)
today_date_kor = datetime.datetime.now().strftime("%Y년 %-m월 %-d일")

# 요청 데이터 설정 (당일 가장 많이 검색된 이슈 요청)
payload = {
    "model": "sonar-pro",
    "max_tokens": 6000,
    "temperature": 0.4,
    "messages": [
        {
            "role": "system",
            "content": (
                "당일 사람들이 가장 많이 검색하거나 본 이슈를 분석하여 제공하세요. "
                "특정 카테고리(정치, 경제, 연예 등)로 구분하지 말고, 검색량과 관심도가 높은 이슈를 중심으로 나열하세요."
                "객관적인 내용을 유지하면서, 논리적으로 정리하세요."
            ),
        },
        {
            "role": "user",
            "content": (
                f"{today_date_kor} 현재, 사람들이 가장 많이 검색하거나 본 이슈 4~5개를 정리해 주세요.\n\n"
                "**조건:**\n"
                "✅ 그 중 최소 2개 이상은 반드시 **한국 국내 이슈**여야 합니다.\n"
                "✅ 나머지는 해외 이슈여도 좋습니다.\n\n"
                "**출력 형식:**\n"
                "1. 이슈 제목\n"
                "이슈에 대한 개요 설명 (3~4줄)\n"
                "- 추가 설명 (해당 이슈에 대한 더 자세한 설명, 관련된 인물이나 단체, 이 이슈에 대한 여론등을 자세히 제시) (4~5줄)\n\n"
                "🔹 특정 주제에 국한하지 말고, 가장 많이 조회된 이슈를 기반으로 정리해 주세요.\n"
                "🔹 출처 표시는 하지 말고 자연스럽게 설명해 주세요."
                "🔹 이슈가 과거의 일과 연관되어 있으면 과거의 이슈까지 곁들여서 설명해주세요."
                "🔹 만약 해외 이슈이거나 국내 정치 이슈이면 관련된 인물이나 관련된 사건까지 5줄 이상으로 최대한 자세히 설명해주세요. 해외 이슈라면 어느 국가에 관한 건지 명확히 표기해주세요"
            ),
        },
    ],
    #"response_format": {"type": "text"},
}

# API 요청 보내기
headers = {"Authorization": f"Bearer {API_KEY}"}
response = requests.post(url, headers=headers, json=payload)

# ✅ API 응답을 확인하고 출력 및 파일 저장
output_text = ""

if response.status_code == 200:
    response_json = response.json()
    if "choices" in response_json:
        output_text = response_json["choices"][0]["message"]["content"]
    elif "message" in response_json:
        output_text = response_json["message"]["content"]
    elif "error" in response_json:
        output_text = f"API 오류 발생: {response_json['error']['message']}"
    else:
        output_text = "예상치 못한 응답 형식: " + str(response_json)
else:
    output_text = f"HTTP 오류 발생: {response.status_code}, 응답 내용: {response.text}"

# ✅ 출력값을 터미널에 표시 (기존 기능 유지)
print(output_text)

# ✅ 파일로 저장 (출력값을 그대로 저장)
with open("output.txt", "w", encoding="utf-8") as file:
    file.write(output_text)

print("✅ API 응답이 `output.txt` 파일에 저장되었습니다.")
```

```python
from elevenlabs import stream
from elevenlabs import VoiceSettings
from elevenlabs.client import ElevenLabs
from IPython.display import Audio
import io
from pydub import AudioSegment

# 🔹 output.txt 파일에서 텍스트 읽기
with open("output.txt", "r", encoding="utf-8") as file:
    text_content = file.read().strip()  # 파일 내용을 읽어서 문자열로 저장

# 🔹 ElevenLabs 클라이언트 초기화
client = ElevenLabs(
    api_key='sk_9fa57e34ff2309975951078428a10145b0ec3fb425bd642b',
)

# 🔹 스트리밍 TTS 변환 요청
audio_stream = client.text_to_speech.convert_as_stream(
    text=text_content,
    voice_id="3MTvEr8xCMCC2mL9ujrI",  # Voice ID 확인 필수
    model_id="eleven_multilingual_v2",  # 올바른 모델 ID 사용
    voice_settings=VoiceSettings(
        speed=1.2
    )
)

# 🔹 오디오 데이터를 하나의 변수로 저장
audio_bytes = b"".join(chunk for chunk in audio_stream if isinstance(chunk, bytes))

# 🔹 Colab에서 바로 재생
audio_display = Audio(audio_bytes, rate=44100, autoplay=True)

# 🔹 WAV 파일로 저장
output_path = "output.wav"
with open(output_path, "wb") as f:
    f.write(audio_bytes)

# 🔹 MP3 -> WAV 변환 (필요 시)
audio_segment = AudioSegment.from_file(output_path, format="mp3")
audio_segment.export(output_path, format="wav")

print("✅ TTS 음성이 output.wav 파일로 저장되었습니다.")

# 🔹 Colab에서 다운로드 (필요하면 실행)
# from google.colab import files
# files.download(output_path)

# 🔹 오디오 플레이어 반환 (Colab에서 재생)
audio_display
```
